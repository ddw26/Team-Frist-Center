### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
url = "https://raw.githubusercontent.com/ddw26/Team-Vand/circle_role_crkaide/output_data/df_values.csv"
catch_df = pd.read_csv(url)
catch_df.head(5)
```

```python
# Drop the non-beneficial ID columns, 'date' and 'fish avail for inspection'.
catch_df = catch_df.drop(['Date file created','Number of fish avail for inspection'],1)
```

```python
# Determine the number of unique values in each column.
catch_df.nunique()
```

```python
# Generate our categorical variable lists
fish_category = catch_df.dtypes[catch_df.dtypes == 'object'].index.tolist()
fish_category
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(catch_df[fish_category]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(fish_category)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
catch_df = catch_df.merge(encode_df,left_index=True,right_index=True)
catch_df = catch_df.drop(columns = fish_category)
catch_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
Y = catch_df['Catch fish_Yes'].values
X = catch_df.drop(["Catch fish_Yes"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,Y,random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 100
hidden_nodes_layer2 = 75
hidden_nodes_layer3 = 50

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1,input_dim=number_input_features,activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2,input_dim=number_input_features,activation='sigmoid'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
#create checkpoints
import os
from tensorflow.keras.callbacks import ModelCheckpoint

#Establish checkpoint mapping
os.makedirs('checkpoints/',exist_ok=True)
checkpoint_path = 'checkpoints/weights.{epoch:02d}.hdf5'
```

```python
#improve accuracy attempt #1
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])

#Save model weights every 5 epochs
callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    save_freq=500)

#Train model
fit_model = nn.fit(X_train_scaled,y_train,epochs=200,callbacks=[callback])

#Review accuracy
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test,verbose=2)
print(f'Loss: {model_loss}, Accuracy: {model_accuracy}')
```

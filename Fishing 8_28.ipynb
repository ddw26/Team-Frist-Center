### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
url = "https://raw.githubusercontent.com/ddw26/Team-Vand/circle_role_crkaide/output_data/df_keys.csv"
catch_df = pd.read_csv(url)
catch_df.head(5)
```

```python
# Create a new DataFrame that holds only the fish available names.
FishAvail = catch_df.filter(['Number of fish avail for inspection'], axis=1)
FishAvail
```

```python
# Drop the 'catch fish column thing' column since it's not going to be used on the clustering algorithm.
catch_df = catch_df.drop('Number of fish avail for inspection', axis=1)
catch_df
```

```python
# Use get_dummies() to create variables for text features.
catch = pd.get_dummies("Number of fish avail for inspection")
dummy = pd.get_dummies("Hours on boat","Kind of day","Fish target species")
combined = pd.concat([catch,dummy],axis=1)
df = catch_df.merge(combined,left_index = True,right_index = True)
df = df.drop('Number of fish avail for inspection',axis = 1)
df2 = df.drop("Kind of day",axis = 1)
df3 = df2.drop("Fish target species", axis=1)
df3
```

```python
#drop date
df3 = df3.drop('Date file created', axis=1)
df3
```

```python
# Standardize the data with StandardScaler().
df_scaled = StandardScaler().fit_transform(df3)
print(df_scaled)
```

```python
# Using PCA to reduce dimension to three principal components. - convert numpy to pandas
pca = PCA(n_components=1)
df_pca = pca.fit_transform(df_scaled)
#df_pca
```
